```python
import xarray as xr
import numpy as np

# List of files to check
files = [

    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/hgrid.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/vcoord.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/bathymetry_mod.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/init_tracers_modified.nc",
    #"/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/obgc_obc.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/init_eta.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/init_vel.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/grid_spec.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/land_mask.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/ocean_mosaic.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/access-rom3-ESMFmesh.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/access-rom3-nomask-ESMFmesh.nc",
    "/scratch/nm03/ee8016/regional_mom6_configs/Ningaloo_domain_02/dust_DS.nc",
]

for f in files:
    try:
        ds = xr.open_dataset(f)
        nan_report = {}
        for v in ds.data_vars:
            arr = ds[v].values
            if np.isnan(arr).any():
                nan_count = np.isnan(arr).sum()
                nan_report[v] = int(nan_count)
        if nan_report:
            print(f"\n⚠️ NaNs found in {f}:")
            for var, count in nan_report.items():
                print(f"   {var}: {count} NaNs")
        else:
            print(f"✅ No NaNs in {f}")
        ds.close()
    except Exception as e:
        print(f"❌ Could not open {f}: {e}")
```
